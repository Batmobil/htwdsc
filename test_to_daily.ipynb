```python
test_dates = pd.date_range('2015-11-01', '2015-11-30')
test_dates = pd.to_datetime(test_dates, format='%d.%m.%Y')
test_dates
```

```python
test_dates = pd.DataFrame(pd.Series(34, index=test_dates), index=test_dates).reset_index()
test_dates.head()
```

```python
test['key'] = 0
test_dates['key'] = 0

test_daily = test_dates.merge(test, how='outer').drop('key', axis=1)
test_daily = test_daily.rename(columns={'index': 'date_', 0: 'date_block_num'}).drop('ID', axis=1)
test_daily.head()
```

```python
test_daily['date'] = test_daily['date_'].apply(lambda d: str(d))
test_daily.head()
```
