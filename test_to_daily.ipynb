```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
train = pd.read_csv('sales_train_v2.csv')
train.head()
```

```python
test = pd.read_csv('test.csv')
test.head()
```

```python
train.tail()
```

```python
test_dates = pd.date_range('2015-11-01', '2015-11-30')
test_dates = pd.to_datetime(test_dates, format='%d.%m.%Y')
test_dates
```

```python
test_dates = pd.DataFrame(pd.Series(34, index=test_dates), index=test_dates).reset_index()
test_dates.head()
```

```python
test['key'] = 0
test_dates['key'] = 0
test_daily = test_dates.merge(test, how='outer').drop('key', axis=1)
test_daily = test_daily.rename(columns={'index': 'date_', 0: 'date_block_num'}).drop('ID', axis=1)
test_daily.head()
```

```python
test_daily['date'] = test_daily['date_'].apply(lambda d: d.strftime('%d-%m-%Y'))
test_daily.head()
```

```python
test_daily = test_daily[['date', 'date_block_num', 'shop_id', 'item_id']]
test_daily.to_csv('test_daily.csv', index=False)
```

```python
test_daily['item_cnt_day'] = pd.Series(np.nan, index=test_daily.index)
test_daily.head()
```

```python
train_test = pd.concat([train, test_daily]).reset_index()
```

```python
train_test.to_csv('train_test.csv', index=False)
```
